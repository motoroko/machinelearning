```
#Categorical Encoding

def delduplicate(a):
  b = a
  return list(Counter(b))

neighbourhood_group,neighbourhood,room_type=[],[],[]

neighbourhood_group = delduplicate(df_training['neighbourhood_group'])
neighbourhood = delduplicate(df_training['neighbourhood'])
room_type = delduplicate(df_training['room_type'])

for i in range(len(df_training)):
  for j in range(len(neighbourhood_group)):
    if df_training['neighbourhood_group'].iloc[i] == neighbourhood_group[j]:
      df_training['neighbourhood_group'].iloc[i] = j+1
      break
  for j in range(len(neighbourhood)):
    if df_training['neighbourhood'].iloc[i] == neighbourhood[j]:
      df_training['neighbourhood'].iloc[i] = j+1
      break
  for j in range(len(room_type)):
    if df_training['room_type'].iloc[i] == room_type[j]:
      df_training['room_type'].iloc[i] = j+1
      break
```

```
a = df_training['last_review'].str.split('-',expand=True)
df_training["tahun"] = a[0].astype(int)
df_training["bulan"] = a[1].astype(int)
df_training['tanggal'] = a[2].astype(int)
df_training
```
